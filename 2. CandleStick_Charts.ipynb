```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

from datetime import datetime
import pandas_datareader.data as data
import yfinance as yf
yf.pdr_override()

import plotly.graph_objects as go
```

**Loading Stock Data**

```python
## Loading SBI Stock Data
start = '2010-01-01'   ## Jan 01, 2010
end = '2020-12-31'     ## Dec 31, 2020

def loadStockData(stock, start, end):
  sbi = data.get_data_yahoo(stock, start=start, end = end)
  return sbi 


sbi = loadStockData('SBIN.NS', start, end)
sbi.head()
```

**Function for Candle Stick Plotting**

```python

def showCandleStick(period_data):
  figure = go.Figure(
      data = go.Candlestick(
          
          x = period_data.index, 
          low = period_data['Low'], 
          high = period_data['High'], 
          close = period_data['Close'], 
          open = period_data['Open'], 
          increasing_line_color = 'green', 
          decreasing_line_color = 'red'
      )
  )
  figure.show()
```

**Smaller Period**

```python
showCandleStick(sbi.loc['2010-01-04':'2010-01-10'])
```

**Longer Period**

```python
showCandleStick(sbi.loc['2010-01-04':'2011-01-04'])
```
